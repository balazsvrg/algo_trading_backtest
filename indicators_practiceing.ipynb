```python
"""
ADX(Avarage Directional Index) - Technical Indicator
"""


import pandas as pd
import numpy as np


otp_data = pd.read_csv('data.csv')
otp_data = otp_data.set_index(pd.DatetimeIndex(otp_data['Date'].values))
print(otp_data)

high = otp_data['High']
low = otp_data['Low']
close = otp_data['Close']
span = 14
todays_high = []
todays_low  = []
todays_close = []
yesterdays_high = []
yesterdays_low = []
yesterdays_close = []
pdms_list = []
ndms_list = []
trs_list = []
ranges = []

for i in range(0, len(otp_data)):
    todays_high.append(high[i])
    todays_low.append(low[i])
    todays_close.append(low[i])
    yesterdays_high.append(high[i-1])
    yesterdays_low.append(low[i-1])
    yesterdays_close.append(close[i-1])
    ranges.append([todays_high[i] - todays_low[i], abs(todays_high[i] - yesterdays_close[i]), abs(todays_low[i] - yesterdays_close[i])])
print('Highs:')
print(todays_high, yesterdays_high)
print('Lows:')
print(todays_low, yesterdays_low)
print('Closes:')
print(todays_close, yesterdays_close)
print('Ranges:')
print(ranges)  


for i in range(0, len(ranges)):
    trs_list.append(max(ranges[i]))
    

for i in range(0, len(todays_high)):
    if i == 0:
        ndms_list.append(0)
    pdm = todays_high[i] - yesterdays_high[i] 
    pdms_list.append(pdm)

for i in range(0, len(yesterdays_low)):
    if i == 0:
        ndms_list.append(0)
    ndm = yesterdays_low[i] - todays_low[i]
    ndms_list.append(ndm)
   


for i in range (0, len(pdms_list)):
    if pdms_list[i] > ndms_list[i]:
        if pdms_list[i] < 0:
            pdms_list[i] = 0
        ndms_list[i] = 0
    if pdms_list[i] < ndms_list[i]:
        if ndms_list[i] < 0:
            ndms_list[i] = 0
        pdms_list[i] = 0

print('This is positive directional movement')
print(pdms_list)
print('This is negative directional movement')
print(ndms_list)
print('This is true ranges:')
print(trs_list)

"""
pdm = todayshigh - yestardayshigh  
ndm = todayslow - yesterdaylow
"""










```
